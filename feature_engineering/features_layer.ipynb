Importando bibliotecas

```python
from gensim.models import CoherenceModel

import pandas as pd
```

Carregando modelo Doc2vec de reclamações

```python
model = CoherenceModel.load('../model/doc2vec_telecom_pandemic_claims')
```

Lendo base de dados de reclamações tratadas

```python
rfzd_claims = pd.read_parquet('../dataset/refined/rfzd_telecom_pandemic_claims.parquet.snappy')
```

```python
rfzd_claims.set_index('id', inplace=True)
```

Inferindo embeddings para cada reclamação

```python
rfzd_claims['doc_embeddings'] = rfzd_claims['tokens_customer_claim'].apply(lambda tokens: model.infer_vector(tokens))
```

Ajustamento dataset para conter apenas o ID e os embeddings

```python
frzd_claims = pd.DataFrame(rfzd_claims['doc_embeddings'].tolist(), index=rfzd_claims.index)
```

Salvando dataset de embeddings como parquet

```python
frzd_claims.to_parquet(
    path='../dataset/featured/frzd_telecom_pandemic_claims.parquet.snappy',
    compression='snappy',
    index=True,
)
```
